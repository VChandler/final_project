```python
import pandas as pd
```

```python
df = pd.read_csv('SpotifyFeatures.csv')
df.head()
```

```python
# If popularity is 50 or more, hit column = 1; else 0
df.loc[df['popularity'] >= 50, 'hit'] = 1
df.loc[df['popularity'] < 50, 'hit'] = 0
df.head()
```

```python
# Drops artist name, track name, track id, duration, time signature
df.drop(['artist_name', 'track_name', 'track_id', 'duration_ms', 'time_signature'], axis = 1, inplace=True)
df.head()
```

```python
# Check number of genres
genre_count = df.genre.value_counts()
genre_count
```

```python
df2 = df[df['genre'] != 'A Capella']
df2.head()
```

```python
# dropping genre, key, popularity, and mode for first pass of data
df2.drop(['genre', 'key', 'mode', 'popularity'], axis = 1, inplace=True)
df2.head()
```

```python
# Separate the features from the target
y = df2['hit']
X = df2.drop(columns='hit')
```

```python
# Split data into testing and training
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
X_train.shape
```

```python
# Creating logistic regression model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                               max_iter=100,
                               random_state=1)
```

```python
# fit model using training data
classifier.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = classifier.predict(X_test)
results = pd.DataFrame({'Prediction': y_pred, 'Actual': y_test}).reset_index(drop=True)
results.head(20)
```

```python
# Get accuracy score
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```
