```python
import pandas as pd
```

```python
songs_for_tableau_df = pd.read_csv('songs_for_tableau.csv')
songs_for_tableau_df.head()
```

```python
tracks_for_tableau_df = pd.read_csv('tracks_for_tableau.csv')
tracks_for_tableau_df.head()
```

```python
joined_df = pd.merge(songs_for_tableau_df, tracks_for_tableau_df, on='track_id', how='inner')
joined_df.head()
```

```python
joined_df.to_csv("joined_df.csv", index = False)
```
