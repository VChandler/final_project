```python
import pandas as pd
```

```python
df = pd.read_csv('SpotifyFeatures.csv')
df.head()
```

```python
# If popularity is 50 or more, hit column = 1; else 0
df.loc[df['popularity'] >= 50, 'hit'] = 1
df.loc[df['popularity'] < 50, 'hit'] = 0
df.head()
```

```python
# Check number of genres
genre_count = df.genre.value_counts()
genre_count
```

```python
# Drops a capella, comedy, anime and Children’s Music
df2 = df[~(df.genre.isin(['A Capella', 'Comedy', 'Children’s Music', 'Anime']))]
```

```python
# Clusters genres
df2.loc[(df['genre'] == 'Hip-Hop') | (df['genre'] == 'R&B') | (df['genre'] == 'Rap') | (df['genre'] == 'Soul'), 'genre2'] = 'Hip_Hop_R&B'
df2.loc[(df['genre'] == 'Movie') | (df['genre'] == 'Soundtrack'), 'genre2'] = 'Cinema'
df2.loc[(df['genre'] == 'Jazz') | (df['genre'] == 'Blues'), 'genre2'] = 'Jazz_Blues'
df2.loc[(df['genre'] == 'Folk') | (df['genre'] == 'Country'), 'genre2'] = 'Folk_Country'
df2.loc[(df['genre'] == 'Electronic') | (df['genre'] == 'Dance'), 'genre2'] = 'Danceclub'
df2.loc[(df['genre'] == 'World') | (df['genre'] == 'Reggaeton'), 'genre2'] = 'World'
df2.loc[(df['genre'] == 'Rock') | (df['genre'] == 'Alternative'), 'genre2'] = 'Rock'
df2.loc[(df['genre'] == 'Classical') | (df['genre'] == 'Opera'), 'genre2'] = 'Classical'
df2.loc[(df['genre'] == 'Ska') | (df['genre'] == 'Reggae'), 'genre2'] = 'Ska_Reggae'
df2.loc[(df["genre"] == "Children's Music"), "genre2"] = "Childrens"
df2.loc[(df['genre'] == 'Pop'), 'genre2'] = 'Pop'
df2.loc[(df['genre'] == 'Indie'), 'genre2'] = 'Indie'

# Check number of genres
new_genre_count = df2.genre2.value_counts()
new_genre_count
```

```python
df2.head()
```

```python
# dropping old genre, key, popularity, and mode
df2.drop(['genre', 'key', 'mode', 'popularity'], axis = 1, inplace=True)
df2.head()
```

```python
# rename genre2 to genre
df2.rename(columns = {'genre2': 'genre'}, inplace = True)
```

```python
# Creates dataframe with artist, track name and id
track_df = df2[['artist_name', 'track_name', 'track_id']].copy()
track_df.head()
```

```python
# Drops artist name, track name, duration, time signature
df2.drop(['artist_name', 'track_name', 'duration_ms', 'time_signature'], axis = 1, inplace=True)
df2.head()
```

```python
# Create sqlalchemy
from sqlalchemy import create_engine
engine = create_engine(f"sqlite:///spotify.sqlite")
```

```python
with engine.connect() as conn:
    df2.to_sql('songs', conn, index=False, if_exists='replace')
    track_df.to_sql('trackinfo', conn, index=False, if_exists='replace')
```
